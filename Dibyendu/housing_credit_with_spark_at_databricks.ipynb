## DATA ##

```python
dataset = spark.table("application_train_csv")
display(dataset)
```

```python
cols = dataset.columns
cols
```

```python
numericalCols = [
 'EXT_SOURCE_3',
 'EXT_SOURCE_2',
 'EXT_SOURCE_1',
 'DAYS_BIRTH',
 'REGION_RATING_CLIENT_W_CITY',
 'DAYS_LAST_PHONE_CHANGE',
 'DAYS_ID_PUBLISH',
 'REG_CITY_NOT_WORK_CITY',
 'FLAG_EMP_PHONE',
 'REG_CITY_NOT_LIVE_CITY',
 'FLAG_DOCUMENT_3',
 'FLOORSMAX_AVG']

categoricalCols = [
'NAME_EDUCATION_TYPE',
'NAME_INCOME_TYPE']
```

```python
featureCols = categoricalCols + numericalCols
```

```python
labelCol = ["TARGET"]
```

```python
datasetSelect = dataset.select(featureCols+labelCol)
display(datasetSelect)
```

```python
datasetSelect
```

```python
import pandas as pd
```

```python
pdf = datasetSelect.select("*").toPandas()
```

```python
pdf.info()
```

```python
for c in pdf.columns[2:-1]:
  pdf[c] = pdf[c].astype(float)
  
pdf.info()
```

```python
datasetSelect = spark.createDataFrame(pdf)
```

```python
datasetSelect
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(inputCols=numericalCols,
                 outputCols=[c+"Imputed" for c in numericalCols])
imputerModel = imputer.fit(datasetSelect)
datasetSelectImputed = imputerModel.transform(datasetSelect)
```

```python
datasetSelectImputed.columns
```

```python
numericalColsImputed = [c for c in datasetSelectImputed.columns 
                    if c.endswith("Imputed")]
numericalColsImputed
```

```python
datasetSelectImputed = datasetSelectImputed.select(numericalColsImputed+
                                                  categoricalCols+
                                                  labelCol)
```

```python
datasetSelectImputed.columns
```

```python
from pyspark.ml.feature import StringIndexer, \
                               OneHotEncoderEstimator, \
                               VectorAssembler
```

```python
stages = []
for col in categoricalCols:
  indexer = StringIndexer(inputCol=col, 
                          outputCol=col+"Index",
                          handleInvalid="keep")
  encoder = OneHotEncoderEstimator(inputCols=[indexer.getOutputCol()],
                                   outputCols=[col+"classVec"])
  stages += [indexer, encoder]
```

```python
indexer = StringIndexer(inputCol="TARGET", outputCol="label")
stages += [indexer]
```

```python
assemblerInput = [c+"classVec" for c in categoricalCols] + numericalColsImputed
assembler = VectorAssembler(inputCols=assemblerInput,
                            outputCol="features")
stages += [assembler]
```

```python
from pyspark.ml import Pipeline
```

```python
pipeline = Pipeline(stages=stages)
pipelineModel = pipeline.fit(datasetSelectImputed)
datasetFinal = pipelineModel.transform(datasetSelectImputed)
```

```python
display(datasetFinal)
```

```python
(trainingData, testData) = datasetFinal.randomSplit([0.7, 0.3], seed= 100)
```

### Logistic regression

```python
from pyspark.ml.classification import LogisticRegression
lr = LogisticRegression(labelCol='label',
                        featuresCol='features',
                        maxIter=10)
lrModel = lr.fit(trainingData)
lrPredictions = lrModel.transform(testData)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
evaluator.evaluate(lrPredictions)
```

```python
evaluator.getMetricName()
```

##### Cross validation

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
lrParamGrid = (ParamGridBuilder()
              .addGrid(lr.regParam, [0.010, 0.10, 1.0, 10.])
              .addGrid(lr.elasticNetParam, [0.0, 0.50, 1.0])
              .build())
```

```python
lrCV = CrossValidator(estimator=lr,
                      estimatorParamMaps=lrParamGrid,
                      evaluator=evaluator,
                      numFolds=5)
```

```python
lrCVModel = lrCV.fit(trainingData)
```

```python
lrCVPredictions = lrCVModel.transform(testData)
```

```python
evaluator.evaluate(lrCVPredictions)
```

```python
print('Model intercept: ', lrCVModel.bestModel.intercept)
```
